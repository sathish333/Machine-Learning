```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'x':['a']*4+['b']*4,'y':[1,0,1,0,1,0,1,1]})
```

```python
class ResponseEncoder:
    def __init__(self):
        self.features={}# stores posterior class probabilities for each value of column(p(y==class)/feature=fi)
        self.default=[]# contains mean for each class label.Useful if we encounter new category while tranforming data
    def fit(self,df,column,response):
        """
        df-> dataframe
        column->  column name that you want to perfrom response encoding
        response-> response column name to check for class labels
        """
        labels=list(df[response].unique())
        for each in df[column].unique():
            temp=df[df[column]==each]
            dict_=dict(temp[response].value_counts())
            feature=[]
            for label in labels:
                if label in dict_:
                    feature.append(dict_.get(label)/len(temp))
                else:
                    feature.append(0)
            self.features[each]=feature

        for label in labels:
            self.default.append(len(df[df[response]==label])/len(df))
    
    def transform(self,series):
        encoded=[]
        for each in series:
            encoded.append(self.features.get(each,self.default))
        

        return np.array(encoded)
    def fit_transform(self,df,column,response):
        self.features={}
        self.default=[]
        self.fit(df,column,response)
        return self.transform(df[column])
            
        
    
```

```python
c=ResponseEncoder()
c.fit(df,column='x',response='y')
```

```python
c.transform(df['x'])
```
